```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm
from matplotlib.gridspec import GridSpec
import numpy as np
```

```python

tanimotos = np.load('/home/jegan/lig_struct_analysis/similarity_rdkit/coxtal_cats_tanimoto.npy')

#plotting the similarity heatmap
fig, ax = plt.subplots(figsize = (6,1.5))
im = ax.imshow(tanimotos, cmap = cm.viridis)

#plt.xticks([149],['OE'])
#for top quartile ligs against themselves
#cbar = plt.colorbar(im)
#cbar.set_label('Pairwise Tanimoto Coefficients', rotation = 270, labelpad= 20)

#for cat_s ligs against the current coxtals
plt.rcParams.update({'font.size': 8})

cbar = plt.colorbar(im, orientation = 'vertical', shrink = 0.4)
cbar.ax.tick_params(labelsize=8)
cbar.set_label('Tanimoto\n coefficient', labelpad= 20, rotation = 270, fontsize = 8)
ax.set_xlabel('CatS Ligands', fontsize = 8)
ax.set_ylabel('Co-crystal\nLigands', fontsize = 8)

#plt.title('Pairwise Tanimoto Comparing CatS to Cocrystal Ligands')
plt.tight_layout()
plt.savefig('/home/jegan/lig_struct_analysis/similarity_rdkit/figs/coxtal_3.png', dpi = 300)
```
